```python
from random import randint
import math
```

```python
def print_base(x, y, z, i):
    print(f"<!-- THE SUSPENDED PLATFORM MODEL {i}-->")
    print(f"<model name=\"landing_platform_{i}\">")
    print(f"\t<include>")
    print(f"\t\t<uri>model://landing_platform</uri>")
    print(f"\t\t<pose>{x} {y} {z} 0 -0 0</pose>")
    print(f"\t</include>")
    print(f"</model>\n")
    
```

```python
list_base = []
num_base  = 0
while(num_base < 20):
    x = randint(43, 46)
    y = randint(-38, 48) # -36 - 48
    z = 2.5
    new_base = [x, y, z]
    add_base = True
    for j in range(len(list_base)):
        distBase = math.sqrt(
            math.pow((new_base[0] - list_base[j][0]), 2) 
            + pow((new_base[1] - list_base[j][1]), 2) 
            + pow((new_base[2] - list_base[j][2]), 2)
            )
        if(distBase < 3):
            add_base = False

    if(add_base or len(list_base) < 1):
        list_base.append(new_base)
        num_base += 1
```

```python
i = 0
for base in list_base:
    print_base(base[0], base[1], base[2], i)
    i += 1
```
